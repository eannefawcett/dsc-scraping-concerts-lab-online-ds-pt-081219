## Write a Function to Scrape all of the Events on the Given Page Events Page

The function should return a Pandas DataFrame with columns for the Event_Name, Venue, Event_Date and Number_of_Attendees.

```python
from bs4 import BeautifulSoup
import requests
import time
base_url = 'https://www.residentadvisor.net'
first_url_extension = '/events'
first_url = base_url + first_url_extension
html_page = requests.get(first_url)
soup = BeautifulSoup(html_page.content, 'html.parser')
soup.prettify
```

```python
next_url_extension = soup.find('li', {'id':'liNext'}).find('a').attrs['href']
next_url = base_url + next_url_extension
next_url
```

```python
concerts = soup.find('div', class_="fl col4")
concerts
```

```python
concerts.findAll('article')
```

```python
# function to call on time
def retrieve_time(soup):
    times = []
    for i in soup.findAll('article'):
        if i.find('time'):
            times.append(i.find('time').text)
        else:
            times.append('N/A')
    return times
retrieve_time(concerts)
```

```python
# function to call time
def retrieve_names(soup):
    names = []
    for i in soup.findAll('article'):
        if i.find('h1').find('a'):
            names.append(i.find('h1').find('a').text)
        else:
            names.append('N/A')
    return names
retrieve_names(concerts)
```

```python
# function to call venue
def retrieve_venue(soup):
    venues = []
    for i in soup.findAll('article'):
        if i.find('h1').find('span').find('a'):
            venues.append(i.find('h1').find('span').find('a').text)
        else:
            venues.append('N/A')
    return venues
retrieve_venue(concerts)
```

```python
# function to call number of attendees
def retrieve_attendees(soup):
    attendees = []
    for i in soup.findAll('article'):
        if i.find('p', {'class':'attending'}):
            attendees.append(i.find('p', {'class':'attending'}).text.split(" ")[0])
        else:
            attendees.append('N/A')
    return(attendees)
retrieve_attendees(concerts)
```

```python
def scrape_events(events_page_url):
    import pandas as pd
    html_page = requests.get(events_page_url)
    soup = BeautifulSoup(html_page.content, 'html.parser')
    concerts = soup.find('div', class_="fl col4")
    time.sleep(5)
    times = retrieve_time(concerts)
    time.sleep(5)
    names = retrieve_names(concerts)
    time.sleep(5)
    venues = retrieve_venue(concerts)
    time.sleep(5)
    attendees = retrieve_attendees(concerts)
    df = pd.DataFrame()
    df['Event_Name'] = names
    df['Venue'] = venues
    df['Event_Date'] = times
    df['Number_of_Attendees'] = attendees
    return df
scrape_events('https://www.residentadvisor.net/events')
```

## Write a Function to Retrieve the URL for the Next Page

```python
def next_page(base_url):
    first_url_extension = '/events'
    first_url = base_url + first_url_extension
    html_page = requests.get(first_url)
    soup = BeautifulSoup(html_page.content, 'html.parser')
    if soup.find('li', {'id':'liNext'}):
        next_url_extension = soup.find('li', {'id':'liNext'}).find('a').attrs['href']
        next_url = base_url + next_url_extension
    return next_url
next_page('https://www.residentadvisor.net')
```

```python
def url_list(base_url, number_to_retrieve):
    url_list = []
    first_url_extension = '/events'
    url_list.append(base_url + first_url_extension)
    for i in range(0, number_to_retrieve):
        time.sleep(5)
        html_page = requests.get(url_list[i])
        soup = BeautifulSoup(html_page.content, 'html.parser')
        try: 
            next_url_extension = soup.find('li', {'id':'liNext'}).find('a').attrs['href']
            url_list.append(base_url + next_url_extension)
        except KeyError:
            break
    return url_list
url_list('https://www.residentadvisor.net', 20)
```

## Scrape the Next 1000 Events for Your Area

Display the data sorted by the number of attendees. If there is a tie for the number attending, sort by event date.

```python
import pandas as pd
urls = url_list('https://www.residentadvisor.net', 20)
times = []
names = []
venues = []
attendees = []
for i in range(0, len(urls)):
    html_page = requests.get(urls[i])
    soup = BeautifulSoup(html_page.content, 'html.parser')
    concerts = soup.find('div', class_="fl col4")
    time.sleep(5)
    times.append(retrieve_time(concerts))
    time.sleep(5)
    names.append(retrieve_names(concerts))
    time.sleep(5)
    venues.append(retrieve_venue(concerts))
    time.sleep(5)
    attendees.append(retrieve_attendees(concerts))
df = pd.DataFrame()
events = []
places = []
dates = []
peoples = []
for week in names:
    for name in week:
        events.append(name)
for week in venues:
    for place in week:
        places.append(place)
for week in times:
    for day in week:
        dates.append(day)
for week in attendees:
    for person in week:
        peoples.append(person)
df['Event_Name'] = events
df['Venue'] = places
df['Event_Date'] = dates
df['Number_of_Attendees'] = peoples
df
```
